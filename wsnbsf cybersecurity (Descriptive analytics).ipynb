# Descriptive Analytics

```python
results = []
for feature in numeric_columns:
    mean = np.mean(df[feature])
    median = np.median(df[feature])
    mode = df[feature].mode()[0]
    results.append({'Feature': feature, 'Mean': mean, 'Median': median, 'Mode': mode})
results_df = pd.DataFrame(results)

results_df
```

```python
summary = []
for feature in numeric_columns:
    feature_values = df[feature]
    feature_range = feature_values.max() - feature_values.min()
    feature_variance = np.var(feature_values)
    feature_stddev = np.std(feature_values)
    feature_coeff_variation = (feature_stddev / feature_values.mean()) * 100
    
    summary.append({
        'Feature': feature,
        'Range': feature_range,
        'Variance': feature_variance,
        'Standard Deviation': feature_stddev,
        'Coefficient of Variation': feature_coeff_variation
    })

summary_df = pd.DataFrame(summary)
summary_df
```

```python
iqr_results = {}
for column in numeric_columns:
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1
    iqr_results[column] = iqr
for column, iqr_value in iqr_results.items():
    print(f"Interquartile Range for {column}: {iqr_value}")
```

```python
from scipy.stats import chi2_contingency
contingency_table = pd.crosstab(df['behaviour'], df['Type'])

print("Contingency Table : \n", contingency_table)

# Calculate marginal totals
contingency_table['Row_Total'] = contingency_table.sum(axis=1)
contingency_table.loc['Col_Total', :] = contingency_table.sum(axis=0)
print("\nContingency Table with Marginal Totals : \n", contingency_table)

# Calculate row and column percentages
row_percentages = contingency_table.div(contingency_table['Row_Total'], axis=0)
col_percentages = contingency_table.div(contingency_table.loc['Col_Total'], axis=1)
print("\nRow Percentages : \n", row_percentages)
print("\nColumn Percentages : \n", col_percentages)

# Measures of association - chi-squared test
chi2, p, dof, expected = chi2_contingency(pd.crosstab(df['behaviour'], df['Type']))
print("\nChi-square statistic: ", chi2)
print("p-value: ", p)
```
